```python
# Shared dataset
shared_data = []

# Binary lock for thread-safe operations
data_lock = threading.Lock()

# Sample data
COURSES = ["Machine Learning", "Cyber Security", "Distributed Systems", "Web Services"]
UNIVERSITIES = ["Stanford University", "MIT", "Harvard University", "UC Berkeley"]
```

```python
def add_entries(thread_id, num_entries=5):
    global shared_data
    for _ in range(num_entries):
        entry = {
            "Year": random.randint(2022, 2024),
            "Course Name": random.choice(COURSES),
            "Grade": random.randint(60, 100),
            "University": random.choice(UNIVERSITIES)
        }
        # Acquire lock before modifying shared dataset
        data_lock.acquire()
        try:
            shared_data.append(entry)
            # Optional: simulate delay
            time.sleep(0.01)
        finally:
            data_lock.release()
```

```python
def simulate_concurrent_addition(num_threads=10, entries_per_thread=5):
    threads = []
    for i in range(num_threads):
        t = threading.Thread(target=add_entries, args=(i, entries_per_thread))
        threads.append(t)
        t.start()

    for t in threads:
        t.join()

    print(f"\n✅ All threads completed. Total entries: {len(shared_data)}")
```

```python
def verify_dataset():
    df = pd.DataFrame(shared_data)
    print("\n📊 Entry Count per Course:")
    print(df['Course Name'].value_counts())
    print("\n🎓 Entry Count per University:")
    print(df['University'].value_counts())
    print("\n🗓️ Entry Count per Year:")
    print(df['Year'].value_counts())
    print("\nAverage Grade:", df['Grade'].mean())
```

```python
import threading
import random
import time
import pandas as pd
from collections import Counter
from google.colab import files

# Upload the file from Google Colab
uploaded = files.upload()
```

```python



# File path of the uploaded file
file_path = '/content/coursegrades (1).txt'


# Read the file into a pandas DataFrame (adjust delimiter if needed)
df = pd.read_csv(file_path, delimiter=",")  # Adjust delimiter if needed (e.g., "\s+" or "\t")

# Print the column names to verify
print("Column names in dataset:", df.columns)

# If the column names are not correct, you can set them manually
df.columns = ['Year', 'Course Name', 'Grade', 'University']

# Show the first few rows to check if it's loaded correctly
print("Dataset Loaded:")
print(df.head())

# Shared dataset for concurrent threads to modify
shared_data = []

# Sample data for adding new entries (if needed)
COURSES = ["Machine Learning", "Cyber Security", "Distributed Systems", "Web Services"]
UNIVERSITIES = ["Stanford University", "MIT", "Harvard University", "UC Berkeley"]

def add_entries(thread_id, num_entries=5):
    global shared_data
    for _ in range(num_entries):
        entry = {
            "Year": random.randint(2022, 2024),
            "Course Name": random.choice(COURSES),
            "Grade": random.randint(60, 100),
            "University": random.choice(UNIVERSITIES)
        }
        # **No Lock Here - For Testing Purposes**
        shared_data.append(entry)
        time.sleep(random.uniform(0.005, 0.02))


def simulate_concurrent_addition(num_threads=10, entries_per_thread=5):
    threads = []
    for i in range(num_threads):
        t = threading.Thread(target=add_entries, args=(i, entries_per_thread))
        threads.append(t)
        t.start()

    for t in threads:
        t.join()

    print(f"\n✅ All threads completed. Total entries: {len(shared_data)}")

# Verify dataset by counting entries and calculating statistics
def verify_dataset():
    # Use the data from the file first (df) and then the shared_data added by threads
    df = pd.DataFrame(shared_data)  # Now this will be your updated dataset

    print("\n📊 Entry Count per Course:")
    print(df['Course Name'].value_counts())

    print("\n🎓 Entry Count per University:")
    print(df['University'].value_counts())

    print("\n🗓️ Entry Count per Year:")
    print(df['Year'].value_counts())

    print("\nAverage Grade:", df['Grade'].mean())

# Main function to execute the process
def main():
    # Start by verifying the dataset that was loaded from the file
    print("Initial dataset statistics (from file):")
    print("\n📊 Entry Count per Course:")
    print(df['Course Name'].value_counts())

    print("\n🎓 Entry Count per University:")
    print(df['University'].value_counts())

    print("\n🗓️ Entry Count per Year:")
    print(df['Year'].value_counts())

    print("\nAverage Grade:", df['Grade'].mean())

    # Now simulate concurrent additions
    simulate_concurrent_addition(num_threads=10, entries_per_thread=5)

    # Verify the final dataset (original + added entries)
    verify_dataset()

# Execute the main function
if __name__ == "__main__":
    main()
```

```python
# Convert shared_data (list of dicts) to a DataFrame
thread_df = pd.DataFrame(shared_data)

# Combine original dataset with the thread-added data
final_df = pd.concat([df, thread_df], ignore_index=True)

# Display final dataset stats
print("✅ Combined Dataset Stats:")
print(final_df.describe(include='all'))
print(f"Total rows after merging: {len(final_df)}")
```
